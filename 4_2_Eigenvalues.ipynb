# 4.1 Eigenvalues
An eigenvector satisfies $A v = \lambda v$  <br>

Now we consider a matrix $V$ in which each column is an eigenvector. And we define the matrix $D$ in which each diagonal entry is a eigenvalue, then:

$A V = V D$

Then, we have: $V^{-1} A V = D$ 

Which can be interpreted as that $D$ is the matrix in eigenvector basis, and $V$ is the change of basis matrix, that goes from the eigenvector basis to the canonical one. 

We are going to consider __Hermitian__ matrices, so that $A$ is always diagonalizable and the eigenvectors are orthonormal. Therefore, we have that $V V^T = V^T V = I$. <br>
So $V$ is a orthogonal matrix. <br><br> 

###### QR decomposition

We want to write a matrix $A$ as $A = Q_1 R_1$ <br>
Where $Q$ is orthogonal and $R$ is upper triangular. <br>

Multiplying by $Q_1^T$, we get that $Q_1^T A = Q_1^T Q_1 R_1 = R_1$ <br>
So we define a New matrix $A_1 = R_1 Q_1$ such that we have: <br>
$A_1 = Q_1^T A Q_1$ <br>
Now we decompose $A_1$ as $A_1 = Q_2 R_2$ and do the same <br>

$A_2 = R_2 Q_2 = Q_2^T A_1 Q_2 = Q_2^T Q_1^T A Q_1 Q_2$



Which is an orthogonal matrix. Then we ave that $D = A_k = V^T A V$. And multiplying on the left by $V$ we find that $A V  = VD$. 



```python
# Calculating eigenstuff with numpy

import numpy as np

A = np.array([[1,2],[2,1]],float)
x,V = np.linalg.eigh(A)
print(x)
print(V)

# or we can calculate the eigenvalsh directly
x = np.linalg.eigvalsh(A)
print(x)
```
